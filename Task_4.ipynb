```python
%matplotlib inline
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.lines import Line2D
```

```python
np.random.seed(42)
```

```python
n_per_group = 120
```

```python
data_A = [
    np.random.normal(loc=60, scale=8, size=n_per_group),
    np.random.normal(loc=55, scale=10, size=n_per_group),
    np.random.normal(loc=70, scale=7, size=n_per_group),
]
```

```python
data_B = [
    np.random.normal(loc=63, scale=9, size=n_per_group),
    np.random.normal(loc=58, scale=9, size=n_per_group),
    np.random.normal(loc=68, scale=8, size=n_per_group),
]
```

```python
categories = ["Group 1", "Group 2", "Group 3"]
```

```python
base_positions = np.arange(1, len(categories) + 1)
offset = 0.15
pos_A = base_positions - offset
pos_B = base_positions + offset
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
```

```python
vp_A = ax.violinplot(data_A, positions=pos_A, widths=0.25,
                     showmeans=True, showmedians=True, showextrema=True)
```

```python
vp_B = ax.violinplot(data_B, positions=pos_B, widths=0.25,
                     showmeans=True, showmedians=True, showextrema=True)
```

```python
ax.set_xticks(base_positions)
ax.set_xticklabels(categories)
ax.set_xlabel("Category")
ax.set_ylabel("Value")
ax.set_title("Multi-Data Series Violin Plot (Side-by-Side)")
ax.grid(True, axis="y", linestyle="--", alpha=0.4)
```

```python
legend_handles = [Line2D([0], [0]), Line2D([0], [0])]
ax.legend(legend_handles, ["Series A", "Series B"], loc="best")
```

```python
plt.show()
```
